**What is Exploratory Data Analysis (EDA)?**

Exploratory Data Analysis (EDA) is an important first step in data science projects. It involves looking at and visualizing data to understand its main features, find patterns, and discover how different parts of the data are connected.

EDA helps to spot any unusual data or outliers and is usually done before starting more detailed statistical analysis or building models

EDA helps to identify hidden patterns and relationships between different data points, which help us in and model building.

**Churn distribution: 26.6 % of customers switched to another firm.**

```python
# Categories
type = ["No", "Yes"]

# Value counts for churn
churn_counts = data['Churn'].value_counts()

# Colors from seaborn
colors = sns.color_palette("pastel")
myexplode = [0, 0.1]
# Create pie chart (with a hole for donut style)
fig, ax = plt.subplots(figsize=(6,6))
wedges, texts, autotexts = ax.pie(
    churn_counts,
    labels=type,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode = myexplode,
    shadow=True,
    wedgeprops=dict(width=0.6)  # <-- donut effect
)

# Central text
plt.text(0, 0, "Churn", ha='center', va='center', fontsize=16, fontweight='bold')
 # ha means horizontal alignment and va means vertical alignment
plt.title("Churn Distributions Percentages", fontsize=16, fontweight="bold")
plt.legend()
plt.show()
```

**Payment Methods:**

Major customers who moved out were having Electronic Check as Payment Method. Customers who opted for Credit-Card automatic transfer or Bank Automatic Transfer and Mailed Check as Payment Method were less likely to move out.

```python
import matplotlib.pyplot as plt
import seaborn as sns

labels = data['PaymentMethod'].unique()
values = data['PaymentMethod'].value_counts()
colors = sns.color_palette("muted", 4)

myexplode = [0.08, 0.08, 0.08, 0.08]
plt.figure(figsize=(7,7))

plt.pie(values,
        labels= labels,
        autopct= '%1.1f%%',
        startangle= 90,
        colors=colors,
        explode = myexplode,
        shadow= True,        
        wedgeprops=dict(width=0.6),
        textprops= {"fontsize":12}
)
plt.text(0, 0, "Payment Mode", ha='center', va='center', fontsize=15, fontweight='bold')
plt.title("Payment Method Distribution", fontsize= 14,  y=1.05, fontweight= "bold")
plt.axis("equal")
plt.legend(loc= "upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt

# crosstab is used to compute a simple cross-tabulation of two (or more) factors or variables in a dataframe
pivot_df = pd.crosstab(data['PaymentMethod'], data['Churn'])

# Plot the stacked bar chart
ax = pivot_df.plot(
    kind='bar',
    stacked=True, # stacked instead of side by side bars
    figsize=(10, 6),
    grid= False,
    color=['gold', 'violet']  # Set colors for 'No' and 'Yes'
)

# Set the title and labels using matplotlib
ax.set_title('Customer Payment Method distribution w.r.t. Churn', fontweight='bold')
ax.set_xlabel('Payment Method')
ax.set_ylabel('Number of Customers')

# Rename the legend title
ax.legend(title='Churn')

# Rotate x-axis labels
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

**Internet services:**

Several customers choose the Fiber optic service and it's also evident that the customers who use Fiber optic have high churn rate, this might suggest a dissatisfaction with this type of internet service. Customers having DSL service are majority in number and have less churn rate compared to Fibre optic service.

```python
labels = ["Female (No-Churn)", "Male (No-Churn)", "Female (Churn)", "Male (Churn)"]
DSL = [965, 992, 219, 240]
Fiber_optic = [889, 910, 664, 633]
NoInternet = [690, 717, 56, 57]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(10,6))

# Bars
ax.bar(x - width, DSL, width, label="DSL") # x - width means to shift the bars to the left
ax.bar(x, Fiber_optic, width, label="Fiber optic") # x means to keep the bars in the center
ax.bar(x + width, NoInternet, width, label="No Internet") # x + width means to shift the bars to the right

# Labels and title
ax.set_xticks(x) # it is used to set the locations of the ticks on the x-axis of the plot
ax.set_xticklabels(labels)
ax.set_xlabel("Customer Segments", fontweight= "bold")
ax.set_ylabel("Count", fontweight= "bold")
ax.set_title("Churn Distribution w.r.t. Internet Service and Gender", fontsize=16, fontweight="bold")
ax.legend()
plt.grid(False)
plt.xticks(rotation=45)
plt.show()
```
